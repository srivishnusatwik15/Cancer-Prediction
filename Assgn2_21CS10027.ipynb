```python
x_train= data_train.drop(columns=['patient', 'cancer'])  
x_test= data_test.drop(columns=['patient', 'cancer']) 
y_train = data_train['cancer']
y_test = data_test['cancer'] 
```

# Machine learning training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import  GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
sca = StandardScaler()
x_train_sca = sca.fit_transform(x_train)
x_test_sca = sca.transform(x_test)


pca = PCA(n_components=35)
x_train_pca=pca.fit_transform(x_train)
x_test_pca=pca.transform(x_test)
```

# Random Forest Classifier

```python
ranfor = RandomForestClassifier(random_state=50)
ranfor.fit(x_train, y_train)
y_pre = ranfor.predict(x_test)

#y_pred
accuracy = accuracy_score(y_test, y_pre)
precision = precision_score(y_test, y_pre)
recall = recall_score(y_test, y_pre)
f1 = f1_score(y_test, y_pre)

print("Accuracy for Random Forest:",accuracy)
print("Precision for Random Forest:", precision)
print("Recall for Random Forest:", recall)

cm_display = ConfusionMatrixDisplay.from_estimator(ranfor, x_test, y_test, display_labels=['ALL', 'AML'])
```

# SVM

```python
linear_svm = SVC(kernel='linear', random_state=42)
polynomial_svm = SVC(kernel='poly', degree=3, random_state=42) 
rbf_svm = SVC(kernel='rbf', random_state=42)
sigmoid_svm = SVC(kernel='sigmoid', random_state=42)


linear_svm.fit(x_train_pca, y_train)
polynomial_svm.fit(x_train_pca, y_train)
rbf_svm.fit(x_train_pca, y_train)
sigmoid_svm.fit(x_train_pca, y_train)


linear_pre = linear_svm.predict(x_test_pca)
polynomial_pre = polynomial_svm.predict(x_test_pca)
rbf_preds = rbf_svm.predict(x_test_pca)
sigmoid_preds = sigmoid_svm.predict(x_test_pca)

linear_accuracy = accuracy_score(y_test, linear_pre)
polynomial_accuracy = accuracy_score(y_test, polynomial_pre)
rbf_accuracy = accuracy_score(y_test, rbf_preds)
sigmoid_accuracy = accuracy_score(y_test, sigmoid_preds)

print("Sigmoid SVM Accuracy:", sigmoid_accuracy)
print("Linear SVM Accuracy:", linear_accuracy)
print("Polynomial SVM Accuracy:", polynomial_accuracy)
print("RBF SVM Accuracy:", rbf_accuracy)

classi = ["Linear SVM", "Polynomial SVM", "RBF SVM", "Sigmoid SVM"]
pre = [linear_pre, polynomial_pre, rbf_preds, sigmoid_preds]

# Create subplots for confusion matrices
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.subplots_adjust(hspace=0.5)

for i, (name, pr) in enumerate(zip(classi, pre)):
    ax = axes[i // 2, i % 2]
    cm = confusion_matrix(y_test, pr)
    seaborn.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, ax=ax)
    ax.set_title(f"Confusion Matrix for {name}")
    ax.set_xlabel("Predicted")
    ax.set_ylabel("Actual")

plt.show()

```

# Neural Networks

```python
param_grid = {
    'hidden_layer_sizes': [(100,), (50, 50), (100, 50, 100)],
    'activation': ['relu', 'tanh', 'logistic'],
    'alpha': [0.0001, 0.001, 0.01],
    'solver': ['lbfgs', 'sgd', 'adam'],
    'max_iter': [200, 300, 400],
}

mlp = MLPClassifier(random_state=42,early_stopping=True)

grid_search = GridSearchCV(mlp, param_grid, cv=5, scoring='accuracy')

grid_search.fit(x_train_pca, y_train)

best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
best_model = grid_search.best_estimator_


y_pre = best_model.predict(x_test_pca)


accuracy = accuracy_score(y_test, y_pre)
print("Best Model Accuracy:", accuracy)

cm_display = ConfusionMatrixDisplay.from_estimator(best_model, x_test_pca, y_test, display_labels=['ALL', 'AML'])
```
