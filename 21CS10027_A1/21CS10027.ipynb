# Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def cost_compute(x, y, theta_value):
    m = len(y)
    h = sigmoid(np.dot(x, theta_value))
    cost = -(1/m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
    return cost

def gradient_descent(x, y, theta_value, learning_rate, epochs):
    m = len(y)
    costs = []
    
    for epoch in range(epochs):
        h = sigmoid(np.dot(x, theta_value))
        gradient = np.dot(x.T, (h - y)) / m
        theta_value -= learning_rate * gradient
        cost = cost_compute(x, y, theta_value)
        costs.append(cost)
    
    return theta_value, costs
```

```python
def logistic_regression(x_train, y_train, x_test):

    x_train = np.c_[np.ones((x_train.shape[0], 1)), x_train]
    x_test = np.c_[np.ones((x_test.shape[0], 1)), x_test]

    theta_value = np.zeros(x_train.shape[1])
    learning_rate = 0.01
    epochs = 1000
    
    theta_value, costs = gradient_descent(x_train, y_train, theta_value, learning_rate, epochs)
    
    predictions = sigmoid(np.dot(x_test, theta_value))
    y_pred = np.round(predictions)
    
    return y_pred
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

y_pre= logistic_regression(x_train, y_train, x_test)
accuracy_logistic = accuracy_score(y_test, y_pre)
precision_logistic = precision_score(y_test,y_pre)
recall_logistic = recall_score(y_test,y_pre)
f1_logistic = recall_score(y_test,y_pre)


print("Logistic Regression Accuracy : ", accuracy_logistic)
print("Logistic Regression Precision : ", precision_logistic)
print("Logistic Regression Recall : ", recall_logistic)
print("Logistic Regression f1-score : ", f1_logistic)
```

# SVM

```python
from sklearn.svm import SVC
```

```python
def svm(x_train, y_train, x_test):

    sca = StandardScaler()
    x_train = sca.fit_transform(x_train)
    x_test = sca.transform(x_test)
    
    svm_model = SVC(kernel='linear', C=1.0)
    svm_model.fit(x_train, y_train)

    y_pre = svm_model.predict(x_test)
    
    return y_pre
```

```python
y_pre = svm(x_train, y_train, x_test)
accuracy_svm = accuracy_score(y_test, y_pre)
precision_svm = precision_score(y_test,y_pre)
recall_svm = recall_score(y_test,y_pre)
f1_svm = recall_score(y_test,y_pre)


print("SVM Accuracy : ", accuracy_svm)
print("SVM Precision : ", precision_svm)
print("SVM Recall : ", recall_svm)
print("SVM f1-score : ", f1_svm) 
```

# Neural Networks

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
def neural_network(x_train, y_train, x_test):
    nn = Sequential()
    nn.add(Dense(16, input_dim=x_train.shape[1], activation='relu'))
    nn.add(Dense(8, activation='relu'))
    nn.add(Dense(1, activation='sigmoid'))

    nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    nn.fit(x_train, y_train, epochs=40, batch_size=32, validation_split=0.2,  verbose=1)

    y_pre =np.round(nn.predict(x_test)).flatten()
    return y_pre
```

```python
y_pre = neural_network(x_train, y_train, x_test)
accuracy_nn = accuracy_score(y_test, y_pre)
precision_nn = precision_score(y_test, y_pre)
recall_nn = recall_score(y_test, y_pre)
f1_nn = f1_score(y_test, y_pre)


print("Neural Networks Accuracy : ", accuracy_nn)
print("Neural Networks Precision : ", precision_nn)
print("Neural Networks Recall : ", recall_nn)
print("Neural Networks f1-score : ", f1_nn) 
```
